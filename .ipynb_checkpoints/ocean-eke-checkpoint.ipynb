```python
def jd_to_dt(jd, ref=0):
    y, m, d, f = jdcal.jd2gcal(jd, ref)
    h = int(f*24)
    return datetime.datetime(y, m, d, h)
jd_to_dt_v = np.vectorize(jd_to_dt)
ds_j1_dt = jd_to_dt_v(ds.j1.values)
ds.j1.values = ds_j1_dt
```

```python
def range_180(longitude):
    if 260 < longitude <= 540:
        longitude = longitude - 360
    elif 540 < longitude <= 650:
        longitude = longitude - 720
    return longitude
range_180_v = np.vectorize(range_180)
ds_lon_180 = range_180_v(ds.lon.values)
ds.lon.values = ds_lon_180
```

```python
ds_A_meters = ds.A.values*0.01
ds.A.values = ds_A_meters
ds_L_meters = ds.L.values*1000
ds.L.values = ds_L_meters
ds_U_m_s = ds.U.values*0.01
ds.U.values = ds_U_m_s
```

```python
df = ds.to_dataframe()
```

```python
df.head(6)
```

* The kinetic energy of a rigid body rotating about a fixed axis

  $E_{k} = \frac{1}{2}I\omega^2$

* The rotational inertia of a cylinder

  $I = \frac{1}{2}mr^{2}$

* Eddy kinetic energy

  $EKE = \frac{1}{4}mv^{2} = \frac{1}{4}\rho (\pi r^{2}h)v^{2} = \frac{1}{4}\pi\rho_{0}h_{0}L^{2}U^{2}$

```python
pi = np.pi
rho_0 = 1021
h_0 = 500
L = df['L']
U = df['U']
```

```python
df['E'] = 0.25*pi*rho_0*h_0*(L**2)*(U**2)
```

```python
df.head()
```

```python
lat = df.lat
lon = df.lon
E = df.E
```

```python
coord = pd.concat([lat, lon, E], axis=1)
```

```python
coord.head()
```

```python
track = df.track
track_count = track.value_counts(normalize=False, sort=True, ascending=True, bins=None, dropna=False).reindex(range(track.min(), track.max()+1))
track_count = pd.DataFrame(track_count)
track_count.columns = ['count']
```

```python
track_count.head()
```

```python
N_track = 245389
N_obs = 22510603
```

```python
lat_o = np.zeros(N_track)
lon_o = np.zeros(N_track)
E_o = np.zeros(N_track)
```

```python
lat_t = np.zeros(N_track)
lon_t = np.zeros(N_track)
E_t = np.zeros(N_track)
```

```python
i = 0
count = 0
while i < N_obs+1:
    count = count + 1
    lat_o[count-1] = coord['lat'][i]
    lon_o[count-1] = coord['lon'][i]
    E_o[count-1] = coord['E'][i]
    i = i + track_count['count'][count]
    lat_t[count-1] = coord['lat'][i-1]
    lon_t[count-1] = coord['lon'][i-1]
    E_t[count-1] = coord['E'][i-1]
```

```python
lat_o_df = pd.DataFrame(lat_o)
lon_o_df = pd.DataFrame(lon_o)
E_o_df = pd.DataFrame(E_o)
```

```python
lat_t_df = pd.DataFrame(lat_t)
lon_t_df = pd.DataFrame(lon_t)
E_t_df = pd.DataFrame(E_t)
```

```python
coord_o = pd.concat([lat_o_df, lon_o_df, E_o_df], axis=1)
coord_o.columns = ['lat', 'lon', 'E']
```

```python
coord_o.head()
```

```python
coord_t = pd.concat([lat_t_df, lon_t_df, E_t_df], axis=1)
coord_t.columns = ['lat', 'lon', 'E']
```
